### Importação
Para inicializar o projeto, realizamos a importação das bibliotecas **Pandas** e **Matplotlib**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

###  Leitura do CSV
Agora, faremos a leitura do arquivo *csv* como *dataset*.

```python
dados = pd.read_csv('googleplaystore.csv')

dados
```

### Atividades

1. Remova as linhas duplicadas

Primeiro conferimos se há dados duplicados:

```python
dados[dados.duplicated()]
```

* Foram encontradas 483 linhas duplicadas. Para removê-las implantaremos uma função:

```python
dados.drop_duplicates(inplace=True)

dados
```

2. Faça um gráfico de barras contendo os top 5 apps por número de instalação

Para essa etapa, usaremos a *Matplotlib*. Como primeiro passo, limparemos a coluna *Installs* e transformaremos os números que são *string* em *int*:

```python
# Removendo a vírgula
dados['Installs'] = dados['Installs'].str.replace(',', '')

# Removendo o sinal de +
dados['Installs'] = dados['Installs'].str.replace('+', '')

# Transformando a string em número
dados['Installs'] = dados['Installs'].astype(int)
```

Podemos ver que houve um erro, por conta de uma linha que provavelmente possui um valor diferente de números, impossibilitando a transformação. Para descobrir qual linha era a causadora, filtrei a linha que possuía o valor de **'Free'**:

```python
dados[dados['Installs'] == 'Free']
```

A partir disso podemos perceber que a linha está com informações incompletas: Não possui categoria, nem gênero, causando inconsistência nos dados.

Para resolver este problema, poderia inserir dados fictícios na linha ou excluí-la.
Optei por excluir pois mesmo que estivesse retirando o app da análise, por ser um app com poucos *installs*, não iria interferir tanto na busca dos dados. Além disso evitaria invenção de dados falsos no caso de inserir informações.

O código de exclusão:

```python
dados = dados.drop(10472)
```

Neste caso, podemos completar a transformação da *string* em número:

```python
# Transformando a string em número
dados['Installs'] = dados['Installs'].astype(int)
```

Partindo para o próximo passo da criação do gráfico de barras, iremos encontrar os apps com maior número de instalações:


```python
top5apps = dados.sort_values('Installs', ascending=False).head(6)
top5apps[['App', 'Installs']]
```

E por fim podemos gerar o gráfico:

```python
plt.figure(figsize=(8,5)) # Tamanho do gráfico 
plt.title("Apps mais instalados") # Título do gráficos
plt.bar(top5apps['App'], top5apps['Installs'] , color='#845EC2') # Apresentação dos dados do gráfico com a cor
plt.xlabel('Nome dos Aplicativos') # Nomeando a label do eixo X
plt.ylabel('Num. de Instalações (Bilhão)') # Nomeando a label do eixo Y
plt.tight_layout() # Mantendo o layout justificado
plt.show() # Apresentando o gráfico
```

Próxima etapa:

3. Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes no dataset de acordo com a frequência em que elas aparecem

Para a criação do gráfico, precisamos novamente separar os dados.
Primeiro fazemos a contagem da frequência das categorias de apps existentes:

```python
dados['Category'].value_counts() # Contagem de categorias
```

Para evitar poluição do gráfico com muitos dados em espaços pequenos, escolhi dividir os dados a serem mostrados em partes. Desta maneira, serão exibidos apenas as 5 principais categorias com maior recorrência, o restante será apresentado como "Outras".

Para isso farei uma nova partição:

```python
contagem = dados['Category'].value_counts() # Variável para Contagem de categorias
maiores_cat = contagem.nlargest(5) # Separando as 5 maiores categorias
outras_cat = pd.Series({'OUTRAS' : contagem[5:].sum()})# Separa as categorias restantes
categorias = pd.concat([maiores_cat, outras_cat]) # Concatenando os dois formatos
categorias # Apresentação das novas categorias
```

Por fim, geramos o gráfico de Pizza:

```python
plt.figure(figsize=(5,5)) # Tamanho do gráfico 
cores= ['#00e1c6','#0040a8', '#006cc8', '#0095d4', '#5d4775', '#2c0073']
plt.title("Principais Categorias de Aplicativos") # Título do gráficos
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', colors=cores) # Apresentação dos dados do gráfico com cores
plt.tight_layout() # Mantendo o layout justificado
plt.show() # Apresentando o gráfico
```

Continuando para a etapa seguinte:

6. Mostre o top 10 apps por número de reviews bem como o respectivo número de reviews. Ordene a lista de forma decrescente por número de reviews.

De início, devemos ter certeza de que os dados da coluna *'Reviews'* são numéricos:

```python
dados['Reviews'] = dados['Reviews'].astype(int) # Transformando string em número caso não esteja
```

Tendo a certeza, podemos fazer a filtragem com a ordenação de forma correta:

```python
dados[['App', 'Reviews']].sort_values('Reviews', ascending=False).head(10) # Filtragem das reviews
```

Podemos ver que houve problema de dualidade na apresentação dos dados. Como solução, utilizei a função *'max'* com *'groupby'* para exibir apenas a entrada com maior número de *reviews* por app:

```python
top10reviews = dados.groupby('App')['Reviews'].max().sort_values(ascending=False).head(10) # Agrupando o top 10 sem dualidades
top10reviews # Apresentando o top 10
```

Chegamos na etapa 7, com o seguinte enunciado:

7. Crie pelo menos mais 2 cálculos sobre o dataset e apresente um em formato de lista e outro em formato de valor.
Por exemplo: "top 10 apps por número de reviews" e "o app mais caro existente no dataset".

Para iniciar, faço a escolha do primeiro cálculo. Apresentarei como lista "top 10 aplicativos com a maior nota de avaliação". Para exibir essa lista, é necessário primeiro repetir o processo de transformar a *string* em número(*float* no caso da avaliação):

```python
dados['Rating'] = dados['Rating'].astype(float)
```

Pós processo de transformações, realizo a filtragem dos aplicativos com maior nota com a seguinte entrada:

```python
top10avaliacao = dados[['App', 'Rating']].sort_values('Rating', ascending=False).head(10)
top10avaliacao
```

Concluído o 1º cálculo, dou seguimento à escolha do 2°: quero encontrar o app mais baixado da categoria comunicação, retornando o nome, a categoria e o número de instalações respectivamente.

Para se concretizar o cálculo, é preciso apenas fazer a filtragem, já que a coluna *Installs* foi convertida nas etapas anteriores.

```python
maisbaixadocomunicacao = dados[dados['Category'] == 'COMMUNICATION'].sort_values('Installs', ascending=False).head(1)
maisbaixadocomunicacao[['App', 'Installs', 'Category']]
```

Concluída esta etapa, chegamos a última:

8. Crie pelo menos outras 2 formas gráficas de exibição dos indicadores acima utilizando a biblioteca Matplotlib.

Conforme dito no enunciado, utilizarei o indicador da etapa 6(Top 10 apps por número de reviews) e depois da etapa 7(Top 10 aplicativos com maior nota de avaliação.)

Para o primeiro, segue a implantação do gráfico:

```python
plt.figure(figsize=(7,4)) # Tamanho do gráfico 
plt.title("Apps com mais Reviews") # Título do gráficos
plt.barh(top10reviews.index[::-1], top10reviews.values[::-1], color='#e66808') # Apresentação dos dados do gráfico com a cor e a ordem decrescente
plt.xlabel('Num. de Reviews(Milhões)') # Nomeando a label do eixo X
plt.ylabel('Nome dos Apps') # Nomeando a label do eixo Y
plt.ticklabel_format(style='sci', axis='x', scilimits=(6,6)) # Exibição em dezena de milhão
plt.tight_layout() # Mantendo o layout justificado
plt.show() # Apresentando o gráfico
```

Para a implantação do último gráfico, a entrada não mudará muito:

```python
plt.figure(figsize=(7,4)) # Tamanho do gráfico 
plt.title("Aplicativos com maior nota de avaliação") # Título do gráficos
plt.barh(top10avaliacao['App'], top10avaliacao['Rating'] , color='#b74582') # Apresentação dos dados do gráfico com a cor
plt.xlabel('Nota de Avaliação') # Nomeando a label do eixo X
plt.ylabel('Nome dos Apps') # Nomeando a label do eixo Y
plt.gca().invert_xaxis() #Inverte o eixo x virando pra esquerda
plt.tight_layout() # Mantendo o layout justificado
plt.show() # Apresentando o gráfico
```
